```python
DATA CLEANING
```

```python
#Columns with Null value Percentage > 35%
```

```python
null_col=app.isnull().sum().sort_values(ascending=False)
null_col=null_col[null_col.values>(0.35*len(app))]
```

```python
plt.figure(figsize=(25,10))
null_col.plot(kind='bar', color='red')
plt.title("Percentage of Missing Values in Application_Data")
plt.ylabel("NULL VALUES PERCENTAGE (NVP)")
plt.xlabel("COLUMNS")
plt.show()
```

```python
#Removing Columns with NVP>35%
app.drop(app.columns[app.isnull().mean()>0.35],axis=1,inplace=True)
app.shape
```

```python
print((round(app.isnull().mean()*100,2).sort_values(ascending=False)).to_string())
```

```python
HANDLING MISSING VALUES
```

```python
#Imputing Mode and Median values for NVP<20%
app.AMT_GOODS_PRICE.fillna(app.AMT_GOODS_PRICE.mode()[0],inplace = True) 
app.NAME_TYPE_SUITE.fillna(app.NAME_TYPE_SUITE.mode()[0],inplace = True) 
app.OBS_30_CNT_SOCIAL_CIRCLE.fillna(app.OBS_30_CNT_SOCIAL_CIRCLE.mode()[0],inplace = True) 
app.DEF_30_CNT_SOCIAL_CIRCLE.fillna(app.DEF_30_CNT_SOCIAL_CIRCLE.mode()[0],inplace = True) 
app.OBS_60_CNT_SOCIAL_CIRCLE.fillna(app.OBS_60_CNT_SOCIAL_CIRCLE.mode()[0],inplace = True)
app.DEF_60_CNT_SOCIAL_CIRCLE.fillna(app.DEF_60_CNT_SOCIAL_CIRCLE.mode()[0],inplace = True)
app.AMT_REQ_CREDIT_BUREAU_YEAR.fillna(app.AMT_REQ_CREDIT_BUREAU_YEAR.mode()[0],inplace = True) 
app.AMT_REQ_CREDIT_BUREAU_QRT.fillna(app.AMT_REQ_CREDIT_BUREAU_QRT.mode()[0],inplace = True)
app.AMT_REQ_CREDIT_BUREAU_MON.fillna(app.AMT_REQ_CREDIT_BUREAU_MON.mode()[0],inplace = True) 
app.AMT_REQ_CREDIT_BUREAU_WEEK.fillna(app.AMT_REQ_CREDIT_BUREAU_WEEK.mode()[0],inplace = True)
app.AMT_REQ_CREDIT_BUREAU_DAY.fillna(app.AMT_REQ_CREDIT_BUREAU_DAY.mode()[0],inplace = True) 
app.AMT_REQ_CREDIT_BUREAU_HOUR.fillna(app.AMT_REQ_CREDIT_BUREAU_HOUR.mode()[0],inplace = True) 
app.EXT_SOURCE_2.fillna(app.EXT_SOURCE_2.median(),inplace = True) 
app.EXT_SOURCE_3.fillna(app.EXT_SOURCE_3.median(),inplace = True) 
```

```python
def credit_group(x):
    if (x < 500000):
        return 'Low'
    elif (x >= 500000 and x < 750000):
        return 'Medium'
    else:
        return 'High'
```

```python
app['CREDIT_GROUP'] = app['AMT_CREDIT'].apply(credit_group)
app.head()
```

```python
def targets_corr(data,title):
    plt.figure(figsize=(10, 5))
    
    mask= np.zeros_like(data)
    mask[np.triu_indices_from(mask)]=True
    with sns.axes_style("white"):
        ax= sns.heatmap(data, mask=mask) 
```

```python
targets_corr(data=t0,title='Correlation for Repayers (m)')
```

```python
#Repayer

Columns=m.columns  #Collecting all Columns of Target0 
corr=m[Columns].corr(method = 'pearson')                        # Finding correlations using corr() func.

corr=corr.where(np.triu(np.ones(corr.shape),k=1).astype(np.bool))

top10_corr0=corr.unstack().reset_index() # Top 10 Correlations           
top10_corr0.columns = ['VAR1','VAR2','CORRELATION']  # dividing Correlated variables to Columns "var1" & "Var2"

top10_corr0.dropna(subset=['CORRELATION'],inplace=True)

top10_corr0['CORR_ABS']=top10_corr0['CORRELATION'].abs()

top10_corr0.sort_values('CORR_ABS', ascending=False).head(10)
```

```python
#Defaulters

Columns=n.columns  #Collecting all Columns of Target0 

corr=n[Columns].corr(method = 'pearson')                        # Finding correlations using corr() func.

corr=corr.where(np.triu(np.ones(corr.shape),k=1).astype(np.bool))

top10_corr0=corr.unstack().reset_index() # Top 10 Correlations           
top10_corr0.columns = ['VAR1','VAR2','CORRELATION']  # dividing Correlated variables to Columns "var1" & "Var2"

top10_corr0.dropna(subset=['CORRELATION'],inplace=True)

top10_corr0['CORR_ABS']=top10_corr0['CORRELATION'].abs()

top10_corr0.sort_values('CORR_ABS', ascending=False).head(10)
```
